```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.DataFrame({'Epoch': range(1, 7),
                   'Training Loss': [0.7468,
                                     0.2834,
                                     0.1260,
                                     0.4980,
                                     0.2897,
                                     0.0916, ],
                   'Validation Loss': [0.7705,
                                       0.6113,
                                       0.7406,
                                       0.2301,
                                       0.2564,
                                       0.2385, ],
                   'Validation Accuracy': [0.7179,
                                           0.8103,
                                           0.8256,
                                           0.9436,
                                           0.9333,
                                           0.9436, ], })
```

```python
df_melted = df.melt(id_vars="Epoch", var_name="Variable", value_name="Value")
# sns.lineplot(x='Epoch', y='Training Loss', data=df)
# sns.lineplot(x='Epoch', y='Validation Loss', data=df)
# sns.lineplot(x='Epoch', y='Validation Accuracy', data=df)

sns.set_theme(style="whitegrid", palette="muted", context="talk")
plt.figure(figsize=(18, 8))
sns.lineplot(data=df_melted, x="Epoch", y="Value", hue="Variable", linewidth=4,marker='o',markersize=14)
plt.rcParams['lines.markersize'] = 100  

# Add labels, title, and legend
plt.title("Training results", fontsize=18)
plt.xlabel("Epoch", fontsize=18)
plt.ylabel("Values", fontsize=18)
plt.legend(title="Metric", fontsize=16)
plt.tight_layout()

plt.show()

df
```
